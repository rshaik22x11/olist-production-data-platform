```python
from pyspark.sql import functions as F

# Source path (Kaggle raw zone)
source_path = "/Volumes/projects/olist/datasets_raw/orders/"

# Target Bronze path
bronze_path = "/Volumes/projects/olist/bronze/orders/"

# Read CSV (NO schema enforcement)
df = spark.read.format("csv") \
    .option("header", "true") \
    .load(source_path)

# Add ingestion timestamp
df = df.withColumn("ingestion_ts", F.current_timestamp())

# Write to Bronze (INTENTIONALLY WRONG)
df.write.format("delta") \
    .mode("append") \
    .save(bronze_path)

print("Bronze load completed")
```
